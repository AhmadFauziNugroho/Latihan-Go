```python
import matplotlib.pyplot as plt

# data
jurusan = [
    'Teknik Informatika',
    'Sistem Informasi',
    'Bisnis Digital',
]

jumlah_mahasiswa = [120, 110, 80]
```

**Basic Bar Chart**

```python
# Basic Bar Chart
plt.bar(jurusan, jumlah_mahasiswa, color = ["Blue", "Orange", "Red"])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa Per Jurusan')
plt.show()
```

**Basic Bar Chart dengan Matplotlib**

```python
# data
jurusan = [
    'Teknik Informatika',
    'Sistem Informasi',
    'Bisnis Digital',
]

jumlah_mahasiswa = [110, 140, 100]

plt.bar(jurusan, jumlah_mahasiswa, color = 'skyblue')
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa Per Jurusan')
plt.xticks(rotation = 45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Kustomisasi Basic Bar Chart**

```python
# data
jurusan = [
    'Teknik Informatika',
    'Sistem Informasi',
    'Bisnis Digital',
]
jumlah_mahasiswa = [90, 100, 130]

# Kustomisasi
colors = ['blue', 'orange', 'red']
plt.bar(jurusan, jumlah_mahasiswa, color = colors)
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Jumlah Mahasiswa Per Jurusan')
plt.xticks(rotation = 45)
for i, v in enumerate(jumlah_mahasiswa):
  plt.text(i, v + 5, str(v), ha='center')
plt.show()
```

**Menambahkan Multiple Bar Chart**

```python
# Data
jurusan = [
    'Teknik Informatika',
    'Sistem Informasi',
    'Bisnis Digital',
]
jumlah_mahasiswa_2020 = [120, 149, 99]
jumlah_mahasiswa_2024 = [110, 140, 100]
x = range(len(jurusan))
width = 0.4

plt.bar(x, jumlah_mahasiswa_2024, width, label='2020', color = "darkred")
plt.bar([p + width for p in x], jumlah_mahasiswa_2020, width=width, label='2024', color= 'skyblue')

plt.title('Jumlah Mahasiswa 2020 vs 2024')
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45)
plt.xticks([p + width / 2 for p in x], jurusan)
plt.legend()
plt.show()
```

**Basic Bar Chart untuk Melihat Pola dan Tren**

```python
# data

jurusan = [
    'Teknik Informatika',
    'Sistem Informasi',
    'Bisnis Digital'
]
tahun = ['2020', '2021', '2022']
data = {
    'Teknik Informatika': [120,149,99],
    'Sistem Informasi': [110,140,100],
    'Bisnis Digital': [90,100,130]
}

width = 0.2
x = range(len(tahun))

plt.bar(x, data['Teknik Informatika'], width=width, label='Teknik Informatika', color='blue')
plt.bar([p + width for p in x], data['Sistem Informasi'], width=width, label='Sistem Informasi', color='orange')
plt.bar([p + width * 2 for p in x], data['Bisnis Digital'], width=width, label='Bisnis Digital', color='Red')

plt.title('Tren Jumlah Mahasiswa per Jurusan 2020-2022')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks([p + width * 1.5 for p in x], tahun)
plt.legend()

plt.show()
```
